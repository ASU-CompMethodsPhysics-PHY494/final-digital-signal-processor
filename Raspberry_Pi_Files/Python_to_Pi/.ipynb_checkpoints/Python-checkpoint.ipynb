```python
from earsketch import *
from Pd import *
from numpy import *
from scipy import *
 
init()
setTempo(101)
 
#Initialise Connections
inp = createUGen(Pd.effect, INPUT)
output = createUGen(Pd.effect, OUTPUT)

connect(input, output)
effect = [None] * len(ampLst)
TIME = [None] * len(ampLst)
VALUE = np.zeros(V)

#Create a new effects
for i in range(0, len(ampLst) - 1):
 

    distortion[i] = Pd.createUGen(reverbEffect, effect)
    reverb[i] = Pd.createUGen(reverbEffect, TIME)
    tremelo[i] = Pd.createUGen(reverbEffect, TIME)
    chorus[i] = Pd.createUGen(reverbEffect, TIME)
 
    #Defaults:
    #Distortion
    setParamMin(distortion[i], TIME, 0)
    setParamMax(distortion[i], TIME, 1000000)
    setParam(distortion[i], TIME, timeList[i])
    
    setParamMin(distortion[i], VALUE, 0)
    setParamMax(distortion[i], VALUE, 5)
    setParam(distortion[i], VALUE, ampList[i])
 

    #Reverb
    setParamMin(reverb[i], TIME, 0)
    setParamMax(reverb[i], TIME, 1000000)
    setParam(reverb[i], TIME, timeList[i])
    
    setParamMin(reverb[i], VALUE, 0)
    setParamMax(reverb[i], VALUE, 5)
    setParam(reverb[i], VALUE, ampList[i])


    #Tremelo
    setParamMin(tremelo[i], TIME, 0)
    setParamMax(tremelo[i], TIME, 1000000)
    setParam(tremelo[i], TIME, timeList[i])
    
    setParamMin(tremelo[i], VALUE, 0)
    setParamMax(tremelo[i], VALUE, 5)
    setParam(tremelo[i], VALUE, ampList[i])
    
    
    #Chorus
    setParamMin(chorus[i], TIME, 0)
    setParamMax(chorus[i], TIME, 1000000)
    setParam(chorus[i], TIME, timeList[i])
    
    setParamMin(chorus[i], VALUE, 0)
    setParamMax(chorus[i], VALUE, 5)
    setParam(chorus[i], VALUE, ampList[i])
    
    
    connect(Pd.effect, inp, distortion[i], 0, 0)
    connect(Pd.effect, reverb[i], TIME[i], 0, 0)
    connect(Pd.effect, tremelo[i], output)
    connect(Pd.effect, chorus[i], output)

finishEffect(Effect)
```
