```python
%matplotlib inline 
x=np.linspace(-np.pi,np.pi,100)
y=np.zeros(100)
for i in range(100): 
    if i%2==0:
         y[i]=.2+np.sin(-np.pi+(2*np.pi)/100*i)
    else: 
        y[i]=-.2+np.sin(-np.pi+(2*np.pi)/100*i)
plt.plot(x,np.sin(x),label='Original Sine')
plt.plot(x,y,label='Distored Sine') 
plt.title("Plain Distorion")
plt.xlabel("x")
plt.ylabel('Amplitude')
plt.legend()
```

```python
import scipy.misc
stair = scipy.misc.ascent()
plt.gray()
plt.imshow(stair)
```

```python
fourier = np.fft.fft2(stair)
fourier_shift = np.fft.fftshift(fourier)
mag = 20*np.log(np.abs(fourier_shift))
plt.subplot(121),plt.imshow(stair, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(mag, cmap = 'gray')
plt.title('Mag'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
mag1=(mag.copy()/100)-2
m2=mag1.copy()[256,:]
m2=np.append(m2,m2)
m2=np.append(m2,m2)

plt.plot(m2,label='stairway')
plt.title("Image Effect")
plt.xlabel("x")
plt.ylabel('Amplitude')
plt.legend()
```
