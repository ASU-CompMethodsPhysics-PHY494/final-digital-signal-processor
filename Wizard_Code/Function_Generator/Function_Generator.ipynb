# Pre-Defined Waves:

```python
#Example Sawtooth Wave
t = linspace(0, 1, num = 1000, endpoint = False)
plt.ylim(-1.5, 1.5)
plt.plot(t, signal.sawtooth((2 * np.pi) * 5 * t))
#Where the 5 controls the frequency of the wave
```

```python
#Example Sine Wave
t = linspace(0, 1, num = 1000, endpoint = False)
plt.ylim(-1.5, 1.5)
plt.plot(np.sin((2 * np.pi) * t))
```

```python
#Example Square Wave
t = linspace(0, 1, num = 1000, endpoint = False)
plt.ylim(-1.5, 1.5)
plt.plot(t, signal.square((2 * np.pi) * 5 * t))
#Where the 5 controls the frequency of the wave
```

```python
#Example Square Wave with modulation
sig = np.sin(2 * np.pi * t)
pwm = signal.square(2 * np.pi * 30 * t, duty=(sig + 1)/2)
plt.plot(t, sig)
plt.plot(t, pwm)
plt.ylim(-1.5, 1.5)

plt.plot(pwm)
```

# In PyAudio:

```python
import pyaudio
import numpy as np

p = pyaudio.PyAudio()

volume = 0.5     # Range [0.0, 1.0]
fs = 44100       # Sampling rate in Hz, must be integer
duration = 1.0   # Seconds, may be a float?
f = 440.0        # Frequency in Hz, may be float?

# Sample Generation, example is sine wave
samples = (np.sin(2*np.pi*np.arange(fs*duration)*f/fs)).astype(np.float32)

# For thw paFloat32, sample values must be in range [-1.0, 1.0]
stream = p.open(format=pyaudio.paFloat32,
                channels=1,
                rate=fs,
                output=True)

# Plays file, could aslo write to file
stream.write(volume*samples)

stream.stop_stream()
stream.close()

p.terminate()
```
